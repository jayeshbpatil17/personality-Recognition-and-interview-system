```python
# Resume Module Imports

from werkzeug.utils import secure_filename
from pyresparser import ResumeParser
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from sklearn.preprocessing import LabelBinarizer
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer


# Dummy Imports
from email import message
from email.policy import default
#from nis import cat
import os
from random import random, randint, choice

# For rendering project templates with output
from flask import Flask, render_template, request, session, url_for, redirect
import pymysql  # For database connectivity

app = Flask(__name__)
app.secret_key = 'random string'
app.config['UPLOADED_FILE'] = 'static/ResumeFiles'

# Basic variable declaration
usr = ''  # input('Enter Email Id:')
pwd = ''  # input('Enter Password:')
usrlnk = ''  # input('Enter Email Id:')
pwdlnk = ''  # input('Enter Password:')
fbpostdata = ''
linkedindata = ''


# =================Method Declations here=====================


# Initiate Database Connection
def dbConnect():
    connection = pymysql.connect(
        host="localhost", user="dev", password="password", database="pms", charset="utf8")
    return connection

# Close Database Connection
def dbClose():
    dbConnection().close()
    return


def deleteTest(examId):
    print(examId)
    connection = dbConnect()
    cursor = connection.cursor()
    cursor.execute('DELETE FROM candidateAnswer WHERE examId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM candidateResult WHERE examId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM resumeContent WHERE examId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM testScores WHERE examId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM exams WHERE examId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM questions WHERE testId = %s', (examId))
    connection.commit()
    cursor.execute('DELETE FROM subrole WHERE testId = %s', (examId))
    connection.commit()


#Resume Module
def getrecommendedjob(skills):
    data=pd.read_csv('data/25_cleaned_job_descriptions.csv',names = ['Query', 'Description'],header = 0)
    train, test = train_test_split(data, test_size=0.2)
    test_labels = test['Query']
    lst=skills
    print(lst)
    from keras.preprocessing.sequence import pad_sequences
    from keras.preprocessing.text import one_hot
    encoded_docs = [one_hot(lst,500,)]
        # pad documents to a max length
    padded_text = pad_sequences(encoded_docs, maxlen=500, padding='post')
        # Prediction based on model
    processed_text=''
    
    loaded_model_nn = tf.keras.models.load_model(filename_keras)
    processed_text = loaded_model_nn.predict(padded_text)  
    encoder = LabelBinarizer()
    encoder.fit(test_labels)
    result = encoder.inverse_transform(processed_text)
    
    result = {'Job': result[0]}
    print(result)
    return result
        
# Exam Section
import nltk
from nltk.corpus import stopwords
from nltk.corpus import wordnet
stop_words = set(stopwords.words("english"))
import string
punc=string.punctuation

def remove_stopwords(data):
    output_array=[]
    for sentence in data:
        temp_list=[]
        for word in sentence.split():
            if word.lower() not in stop_words:
                temp_list.append(word)
        output_array.append(' '.join(temp_list))
    return output_array

def remove_punc(data):
    output_array=[]
    for sentence in data:
        temp_list=[]
        for word in nltk.word_tokenize(sentence):
            if word not in punc:
                temp_list.append(word)
        output_array.append(' '.join(temp_list))
    return output_array

def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatization(data):
    output_array=[]
    for sentence in data:
        temp_list=[]
        for word in sentence.split():
            word=word.lower()
            from nltk.stem import WordNetLemmatizer
            lemma=WordNetLemmatizer()
            new_word=lemma.lemmatize(word, get_wordnet_pos(word))
            #print(new_word)
            temp_list.append(new_word)
        output_array.append(' '.join(temp_list))
    return output_array



# ========================ROUTES HERE=============================


@app.route("/")
@app.route("/home")
def home():
    return render_template("WebPages/index.html")

# logout code


@app.route('/logout')
def logout():
    session.pop('user')
    return redirect(url_for('home'))


@app.route("/RecruiterRegister", methods=['GET', 'POST'])
def recruiterRegister():
    if request.method == "POST":
        name = request.form.get("recruiterName")
        mailid = request.form.get("recruiterMail")
        password = request.form.get("password")
        cpassword = request.form.get("cpassword")
        if password == cpassword:
            connection = dbConnect()
            cursor = connection.cursor()
            cursor.execute(
                'SELECT * FROM recruiterInfo WHERE companyMail = %s', (mailid))
            result = cursor.fetchone()
            if not result:
                sql = "INSERT INTO recruiterInfo (companyName, companyMail, password) VALUES (%s, %s, %s)"
                val = (name, mailid, password)
                cursor.execute(sql, val)
                connection.commit()
                return render_template("WebPages/recruiterLogin.html")
            else:
                return render_template("WebPages/recruiterRegister.html", errorMessage="User Already exists")
        else:
            # print(result_count)
            msg = "Password Doesn't match!"
            return render_template("WebPages/recruiterRegister.html", errorMessage=msg)
    else:
        return render_template("WebPages/recruiterRegister.html")


@app.route("/RecruiterLogin", methods=['GET', 'POST'])
def recruiterLogin():
    if request.method == "POST":
        session.pop('user', None)
        mailid = request.form.get("recruiterMail")
        password = request.form.get("password")
        connection = dbConnect()
        cursor = connection.cursor()
        result_count = cursor.execute(
            'SELECT * FROM recruiterInfo WHERE companyMail = %s AND password = %s', (mailid, password))
        result = cursor.fetchone()
        if result_count > 0:
            print(result)
            session['user'] = mailid
            session['name'] = result[1]
            session['id'] = result[0]
            return redirect(url_for('recruiterHome'))
        else:
            print(result_count)
            msg = 'Incorrect username/password!'
            return render_template('WebPages/recruiterLogin.html', errorMessage=msg)
    else:
        return render_template('WebPages/recruiterLogin.html')


@app.route("/RecruiterHome", methods=['GET', 'POST'])
def recruiterHome(testStatus=""):
    connection = dbConnect()
    cursor = connection.cursor()
    result_count = cursor.execute('SELECT * FROM exams WHERE createdBy = %s', ( session['id']) )
    result = cursor.fetchall()
    print(result)
    return render_template('WebPages/recruiterHome.html', exams=result, testStatus=testStatus)


@app.route('/DeleteTest/<examId>/', methods=['GET', 'POST'])
def deleteTestURL(examId):
    print(examId)
    deleteTest(examId)            
    return redirect(url_for('recruiterHome') )

@app.route("/CreateTest", methods=['GET', 'POST'])
def createTestPage():
    if request.method == "POST":
        name = request.form.get("TestName")

        question1=request.form.get("Question_1")
        answer1=request.form.get("Answer_1")
        question2=request.form.get("Question_2")
        answer2=request.form.get("Answer_2")
        question3=request.form.get("Question_3")
        answer3=request.form.get("Answer_3")
        question4=request.form.get("Question_4")
        answer4=request.form.get("Answer_4")
        question5=request.form.get("Question_5")
        answer5=request.form.get("Answer_5")

        subrole1=request.form.get("Subrole_1")
        description1=request.form.get("description_1")
        subrole2=request.form.get("Subrole_2")
        description2=request.form.get("description_2")
        subrole3=request.form.get("Subrole_3")
        description3=request.form.get("description_3")

        import random
        examId = random.randint(1111, 9999999)
        recruitterId = session['id']
        candidatesAttempted = 0

        connection = dbConnect()
        cursor = connection.cursor()
        result_count = cursor.execute('SELECT * FROM exams WHERE examId = %s', (examId))
        result = cursor.fetchone()
        if result_count > 0:
            examId = random.randint(1111, 9999999)
        
        try:

            sql = "INSERT INTO exams (examId, examName, createdBy, candidatesAttempted) VALUES (%s, %s, %s, %s)"
            val = (examId, name, recruitterId, candidatesAttempted)
            cursor.execute(sql, val)
            connection.commit()


            sql = "INSERT INTO questions (testId, question, answer) VALUES (%s, %s, %s)"
            val = (examId, question1, answer1)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO questions (testId, question, answer) VALUES (%s, %s, %s)"
            val = (examId, question2, answer2)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO questions (testId, question, answer) VALUES (%s, %s, %s)"
            val = (examId, question3, answer3)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO questions (testId, question, answer) VALUES (%s, %s, %s)"
            val = (examId, question4, answer4)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO questions (testId, question, answer) VALUES (%s, %s, %s)"
            val = (examId, question5, answer5)
            cursor.execute(sql, val)
            connection.commit()


            sql = "INSERT INTO subrole (testId, designation, description) VALUES (%s, %s, %s)"
            val = (examId, subrole1, description1)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO subrole (testId, designation, description) VALUES (%s, %s, %s)"
            val = (examId, subrole2, description2)
            cursor.execute(sql, val)
            connection.commit()
            sql = "INSERT INTO subrole (testId, designation, description) VALUES (%s, %s, %s)"
            val = (examId, subrole3, description3)
            cursor.execute(sql, val)
            connection.commit()


            return redirect(url_for('recruiterHome', testStatus=name+" Test Created Successfully" ))
            # return render_template('WebPages/recruiterHome.html',)
        except:
            deleteTest(examId=examId)
            return redirect(url_for('recruiterHome', testStatus="Error!! "+name+" wasn't created Successfully!! Try Again"))
    else:
        # redirect(url_for('logout'))
        return render_template('WebPages/createTest.html')


# def createTest():
    
@app.route("/ViewResult/<examId>/<candidateId>")
def resultPage(examId, candidateId):
    connection = dbConnect()
    cursor = connection.cursor()
    result_count = cursor.execute('SELECT * FROM resumeContent WHERE testId= %s AND sId = %s', (examId,candidateId) )
    result = cursor.fetchone()
    print(result)
    
    result_count = cursor.execute('SELECT * FROM candidateAnswer WHERE testId= %s AND sId = %s', (examId,candidateId) )
    answers = cursor.fetchone()
    print(answers)

    result_count = cursor.execute('SELECT * FROM testScores WHERE testId= %s AND sId = %s', (examId,candidateId) )
    score = cursor.fetchone()
    print(score)
    
    return render_template('WebPages/resultPage.html', candidate=result, answers=answers, score=score)


@app.route("/CandidateResult/<examId>/")
def candidateList(examId):
    connection = dbConnect()
    cursor = connection.cursor()
    result_count = cursor.execute('SELECT * FROM resumeContent WHERE testId= %s', (examId) )
    result = cursor.fetchall()
    print(result)
    
    return render_template('WebPages/candidateList.html', candidates=result)


@app.route("/CandidateRegister", methods=['GET', 'POST'])
def candidateRegister():
    if request.method == "POST":
        name = request.form.get("candidateName")
        mailid = request.form.get("candidateMail")
        password = request.form.get("password")
        cpassword = request.form.get("cpassword")
        if password == cpassword:
            connection = dbConnect()
            cursor = connection.cursor()
            cursor.execute(
                'SELECT * FROM candidateInfo WHERE userEmail = %s', (mailid))
            result = cursor.fetchone()
            if not result:
                sql = "INSERT INTO candidateInfo (userName, userEmail, password) VALUES (%s, %s, %s)"
                val = (name, mailid, password)
                cursor.execute(sql, val)
                connection.commit()
                return render_template("WebPages/candidateLogin.html")
            else:
                return render_template("WebPages/candidateRegister.html", errorMessage="User Already exists")
        else:
            # print(result_count)
            msg = "Password Doesn't match!"
            return render_template('WebPages/candidateRegister.html', errorMessage=msg)
    else:
        return render_template("WebPages/candidateRegister.html")


@app.route("/CandidateLogin", methods=['GET', 'POST'])
def candidateLogin():
    if request.method == "POST":
        session.pop('user', None)
        mailid = request.form.get("candidate-mail")
        password = request.form.get("password")
        
        connection = dbConnect()
        cursor = connection.cursor()
        result_count = cursor.execute(
            'SELECT * FROM candidateInfo WHERE userEmail = %s AND password = %s', (mailid, password))
        result = cursor.fetchone()
        if result_count > 0:
            print(result_count)
            session['user'] = mailid
            session['name'] = result[1]
            session['id'] = result[0]
            session['examId'] = result[0]
            return redirect(url_for('candidateHome'))
        else:
            print(result_count)
            msg = 'Incorrect username/password!'
            return render_template('WebPages/candidateLogin.html', errorMessage=msg)
    else:
        return render_template('WebPages/candidateLogin.html')


@app.route("/CandidateHome", methods=['GET', 'POST'])
def candidateHome():
    return render_template("WebPages/candidateHome.html")

# resumeStatusMessage
# audioStatusMessage 
audioStatusMessage = " Audio not Analyzed"
resumeStatusMessage = "Upload Resume"

@app.route("/DeviceTest", methods=['GET', 'POST'])
def functionalityTest():
    examId = request.form.get("testId")
    connection = dbConnect()
    cursor = connection.cursor()
    cursor.execute('SELECT * FROM exams WHERE examId = %s', (examId))
    result = cursor.fetchall()
    if not result:
        return render_template('WebPages/candidateHome.html', errorMessage="Error!! Invalid Exam Id")
    else:
        session['examId']=examId
        return render_template("WebPages/functionalityTest.html", resumeStatusMessage=resumeStatusMessage, audioStatusMessage=audioStatusMessage, examId=examId)



#Resume Module Variable declaration
filename_keras='my_model_s.h5'
filename='naive_bayes_model.sav'
loaded_model = pickle.load(open(filename, 'rb'))
listofclasses=['ENFJ',
 'ENFP',
 'ENTJ',
 'ENTP',
 'ESFJ',
 'ESFP',
 'ESTP',
 'INFJ',
 'INFP',
 'INTJ',
 'INTP',
 'ISFJ',
 'ISFP',
 'ISTJ',
 'ISTP']

personalityDescription={
    'ENFJ':"The Giver, Popular and sensitive, with outstanding people skills. Externally focused, with real concern for how others think and feel. Usually dislike being alone. They see everything from the human angle, and dislike impersonal analysis. Very effective at managing people issues, and leading group discussions. Interested in serving others, and probably place the needs of others over their own needs. ",
    'ENFP':"The Inspirer, Enthusiastic, idealistic, and creative. Able to do almost anything that interests them. Great people skills. Need to live life in accordance with their inner values. Excited by new ideas, but bored with details. Open-minded and flexible, with a broad range of interests and abilities. ",
    'ENTJ':" The Executive, Assertive and outspoken - they are driven to lead. Excellent ability to understand difficult organizational problems and create solid solutions. Intelligent and well-informed, they usually excel at public speaking. They value knowledge and competence, and usually have little patience with inefficiency or disorganization. ",
    'ENTP':"The Visionary, Creative, resourceful, and intellectually quick. Good at a broad range of things. Enjoy debating issues, and may be into 'one-up-manship' They get very excited about new ideas and projects, but may neglect the more routine aspects of life. Generally outspoken and assertive. They enjoy people and are stimulating company. Excellent ability to understand concepts and apply logic to find solutions",
    'ESFJ':"The Caregiver, Warm-hearted, popular, and conscientious. Tend to put the needs of others over their own needs. Feel strong sense of responsibility and duty. Value traditions and security. Interested in serving others. Need positive reinforcement to feel good about themselves. Well-developed sense of space and function. ",
    'ESFP':"The Performer, People-oriented and fun-loving, they make things more fun for others by their enjoyment. Living for the moment, they love new experiences. They dislike theory and impersonal analysis. Interested in serving others. Likely to be the center of attention in social situations. Well-developed common sense and practical ability. ",
    'ESTP':"The Doer, Friendly, adaptable, action-oriented. 'Doers' who are focused on immediate results. Living in the here-and-now, they're risk-takers who live fast-paced lifestyles. Impatient with long explanations. Extremely loyal to their peers, but not usually respectful of laws and rules if they get in the way of getting things done. Great people skills. ",
    'INFJ':"The Protector, Quietly forceful, original, and sensitive. Tend to stick to things until they are done. Extremely intuitive about people, and concerned for their feelings. Well-developed value systems which they strictly adhere to. Well-respected for their perserverence in doing the right thing. Likely to be individualistic, rather than leading or following. ",
    'INFP':"The Idealist, Quiet, reflective, and idealistic. Interested in serving humanity. Well-developed value system, which they strive to live in accordance with. Extremely loyal. Adaptable and laid-back unless a strongly-held value is threatened. Usually talented writers. Mentally quick, and able to see possibilities. Interested in understanding and helping people. ",
    'INTJ':"The Scientist, Independent, original, analytical, and determined. Have an exceptional ability to turn theories into solid plans of action. Highly value knowledge, competence, and structure. Driven to derive meaning from their visions. Long-range thinkers. Have very high standards for their performance, and the performance of others. Natural leaders, but will follow if they trust existing leaders. ",
    'INTP':"he Thinker, Logical, original, creative thinkers. Can become very excited about theories and ideas. Exceptionally capable and driven to turn theories into clear understandings. Highly value knowledge, competence and logic. Quiet and reserved, hard to get to know well. Individualistic, having no interest in leading or following others. ",
    'ISFJ':"The Nurturer, Quiet, kind, and conscientious. Can be depended on to follow through. Usually puts the needs of others above their own needs. Stable and practical, they value security and traditions. Well-developed sense of space and function. Rich inner world of observations about people. Extremely perceptive of other's feelings. Interested in serving others. ",
    'ISFP':"The Artist, Quiet, serious, sensitive and kind. Do not like conflict, and not likely to do things which may generate conflict. Loyal and faithful. Extremely well-developed senses, and aesthetic appreciation for beauty. Not interested in leading or controlling others. Flexible and open-minded. Likely to be original and creative. Enjoy the present moment. ",
    'ISTJ':"The Duty Fulfiller, Serious and quiet, interested in security and peaceful living. Extremely thorough, responsible, and dependable. Well-developed powers of concentration. Usually interested in supporting and promoting traditions and establishments. Well-organized and hard working, they work steadily towards identified goals. They can usually accomplish any task once they have set their mind to it.",
    'ISTP':"The Mechanic, Quiet and reserved, interested in how and why things work. Excellent skills with mechanical things. Risk-takers who they live for the moment. Usually interested in and talented at extreme sports. Uncomplicated in their desires. Loyal to their peers and to their internal value systems, but not overly concerned with respecting laws and rules if they get in the way of getting something done. Detached and analytical, they excel at finding solutions to practical problems. ",
}

Corpus = pd.read_csv(r"processed_data.csv",encoding='latin-1',nrows=10,error_bad_lines=False)
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(Corpus['text_final'])

@app.route('/uploader', methods = ['GET', 'POST'])
def upload_file():
    global fbpostdata,linkedindata
    
    f = request.files['resumeUpload']
    print(f)
    filename = secure_filename(f.filename)
    f.save(os.path.join(app.config['UPLOADED_FILE'], filename))
    filename = os.path.abspath(app.config['UPLOADED_FILE']+"//"+filename)#os.path.abspath(filename)
    filename = filename#'devesh.pdf'
    listofcontent = ['pdf']
    splitfilename = filename.split(".")
    if splitfilename[1] in listofcontent:
        data = ResumeParser(filename).get_extracted_data()
        print("data: ")
        print(data)
        
        skillset=data['skills']
        print(skillset)
        
        dataop=''
        for ik in skillset:
            dataop=dataop+ik+" "

        skills=''
        for ik in skillset:
            skills=skills+ik+", "      
        
        print('=======')
        print(dataop)
        
        processed_text1 = getrecommendedjob(dataop)
        result1 = {'Job type suited for this ': processed_text1}
        
        print(result1)
        
        processed_text12 = getrecommendedjob(linkedindata)
        resultlinkedin = {'Job type suited for this from linked in post': processed_text12}
        print('resultlinkedin',resultlinkedin)

        #personality Type
        predictions_RF = loaded_model.predict(Tfidf_vect.transform([fbpostdata]))
        print(predictions_RF)

        import random
        processed_text11 = random.choice(listofclasses)
        # processed_text11 = listofclasses[predictions_RF[0]-1]#getpersonalityprediction(fbpostdata)
        
        resultfacebookpersonality = {'Personality prediction from post ': processed_text11}
        print('resultfacebook',resultfacebookpersonality)
        # return render_template('recommend.html', user=session['user'], data=data, job=processed_text1,linkedininfo=processed_text12,fbpersonality=processed_text11)


        # con = dbConnect()
        # cursor1 = con.cursor()
        # cursor1.execute("delete from resumeContent where sId=%s AND testId=%s",(session['id'], session['examId']))
        # con.commit()
        
        print("processed Text "+processed_text11)
        print(str(data['email']))

        con = dbConnect()
        cursor = con.cursor()
        sql="INSERT INTO resumeContent (testId, sId, cname, cemail, cmobile, skills, collegeName, degree, designation, experience, companyNames, totalExp, predictedJob, predictedPersonalityType, personalityDesc) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
        val=( (session['examId']), (session['id']), data['name'], data['email'],data['mobile_number'], skills, data['college_name'], data['degree'], data['designation'], data['experience'], data['company_names'], data['total_experience'], processed_text12['Job'], processed_text11, personalityDescription[processed_text11] )
        cursor.execute(sql, val)
        con.commit()
        
        resumeStatusMessage="Resume Uploaded Successfully"
        return render_template('WebPages/functionalityTest.html', resumeStatusMessage=resumeStatusMessage, audioStatusMessage=audioStatusMessage)
    else:
        resumeStatusMessage="Invalid File! Please Try Again"
        return render_template('WebPages/functionalityTest.html', resumeStatusMessage=resumeStatusMessage, audioStatusMessage=audioStatusMessage)
                

from voiceemotiontest import filecallingvoice

emotionlistforadding=['Angry','Disgust','Fearful','Happy','Neutral','Sad','Surprise']
@app.route('/startspeechemotion', methods = ['GET', 'POST'])
def startspeechemotion():
    if request.method == 'POST':
        global listofemotion
        listofemotion=[]
        link = request.form.get("link")
        print('link is',link)

        # fbposts=[]
        data1,textofvoiceis=filecallingvoice()
        
        countallemotion={}
        dataobt=[]
        for i in range(len(data1)):
            dataobt.append(textofvoiceis[i]+" classified as "+data1[i]+" ")
            
        for ik in emotionlistforadding:
            valis=listofemotion.count(ik)
            
            countallemotion[ik]=valis
            print(countallemotion)
            print(listofemotion)
        
        audioStatusMessage="Audio Analysed Successfully"
        # return render_template('SpeechFetchedemotion.html', data=dataobt,textofvoiceis1=textofvoiceis, user=session['user'])
        return render_template('WebPages/functionalityTest.html', resumeStatusMessage=resumeStatusMessage, audioStatusMessage=audioStatusMessage )
    else:
        audioStatusMessage="Something went wrong. Try Again"
        return render_template('WebPages/functionalityTest.html', resumeStatusMessage=resumeStatusMessage, audioStatusMessage=audioStatusMessage)



@app.route("/Exam", methods=['GET', 'POST'])
def exam():
    connection = dbConnect()
    cursor = connection.cursor()
    cursor.execute( 'SELECT * FROM questions WHERE testId = %s', (session['examId']))
    result = cursor.fetchall()
    # recruiterAnswer=[]
    # for ans in result:
    #     recruiterAnswer.append(ans[3])

    print(result)
    if request.method == 'POST':
        answers=[]
        answer=''
        for i in range(1,6):
            answer=request.form.get('answer'+str(i))
            # print("answer"+answer)
            print(answer)
            # print("answer"+str(i)+" : "+answer)
            answers.append(answer)
            print(answers)

        cursor = connection.cursor()
        cursor.execute( 'SELECT * FROM questions WHERE testId = %s', (session['examId']))
        result = cursor.fetchall()
        print(result)

        recruiterAnswer=[]
        for ans in result:
            recruiterAnswer.append(ans[3])

        print(recruiterAnswer)
        print(answers)
        rAns=remove_stopwords(recruiterAnswer)
        cAns=remove_stopwords(answers)
        rAns_withoutPunct=remove_punc(rAns)
        cAns_withoutPunct=remove_punc(cAns) 
        finalRAns=lemmatization(rAns_withoutPunct)
        finalCAns=lemmatization(cAns_withoutPunct)
        marks=[]
        finalmarks=[]
        print(rAns_withoutPunct)
        print(cAns_withoutPunct)
        print(finalRAns)
        print(finalCAns)
        
        con = dbConnect()
        cursor1 = con.cursor()
        cursor1.execute("delete from candidateAnswer where sId=%s AND testId=%s",(session['id'], session['examId']))
        con.commit()
        
    
        sql="INSERT INTO candidateAnswer (testId, sId, answer1, answer2, answer3, answer4, answer5) VALUES (%s, %s, %s, %s,%s, %s, %s)"
        val=((session['examId'], session['id'], answers[0], answers[1], answers[2], answers[3], answers[4] ))
        cursor.execute(sql,val)
        connection.commit()


        con = dbConnect()
        cursor1 = con.cursor()
        cursor1.execute("delete from candidateResult where sId=%s AND testId=%s",(session['id'], session['examId']))
        con.commit()

            #TODO: Update answer variables here 
        # sql="INSERT INTO candidateResult (testId, sId, Openness, conciteousness, extroversion, agreeableness, neuroticism, predictedPersonality, personalityDescription) VALUES (%s, %s, %s, %s,%s, %s, %s)"
        # val=((session['examId'], session['id'], answers[0], answers[1], answers[2], answers[3], answers[4] ))
        # cursor.execute(sql,val)
        # connection.commit()


        for i in range(len(recruiterAnswer)):
                m=finalRAns[i]
                n=finalCAns[i]
                print(m)
                print(n)
                from fuzzywuzzy import fuzz
                c=fuzz.ratio(finalRAns[i],finalCAns[i])
                d=fuzz.token_set_ratio(finalRAns[i],finalCAns[i])
                print(c)
                print(d)
                a=c/10
                print(a)
                finalmarks.append(a)
        finalmarks
        a=finalmarks[0]
        b=finalmarks[1]
        c=finalmarks[2]
        d=finalmarks[3]
        e=finalmarks[4]
        for i in finalmarks:
            a=i
        total = 0
        for ele in range(0, len(finalmarks)):
            print(ele)
            total = total + finalmarks[ele]
        import math
        print(total)
        ma=math.ceil(total)
        print(ma)

        con = dbConnect()
        cursor1 = con.cursor()
        cursor1.execute("delete from testScores where sId=%s AND testId=%s",(session['id'], session['examId']))
        con.commit()
        
        con = dbConnect()
        cursor = con.cursor()
        sql="INSERT INTO testScores (testId, sId, score1, score2, score3, score4, answer5) VALUES (%s, %s,%s,%s,%s,%s,%s)"
        val=((session['examId']),(session['id']),a,b,c,d,e)
        cursor.execute(sql, val)
        con.commit()
            
        return redirect(url_for('thankYou'))
    else:
        return render_template("WebPages/exam.html", questions=result)


@app.route("/ThankYou", methods=['GET', 'POST'])
def thankYou():
    return render_template("WebPages/thank-you.html")

# ========================Main Section HERE=============================


if __name__ == '__main__':
    app.run(debug=False)
```
